```python
import pandas as pd

# Specify the file path
excel_file = 'sales data.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file)

# Display data types of variables
print("Data types of variables:")
print(df.dtypes)
```

```python


# Define a function to check if a value can be converted to numeric
def is_numeric(value):
    if isinstance(value, (int, float)):
        return True
    elif isinstance(value, str):
        try:
            float(value)
            return True
        except ValueError:
            return False
    else:
        return False


# Dictionary to store non-numeric values for each column
non_numeric_values = {}

# List of columns to iterate over
columns_to_check = ['country', 'order_value_EUR', 'date', 'category', 'customer_name', 'sales_manager', 'sales_rep', 'device_type']

# Iterate over each column
for col in columns_to_check:
    non_numeric_values[col] = []
    # Iterate over each value in the column and check if it's numeric
    for value in df[col]:
        if not is_numeric(value):
            non_numeric_values[col].append(value)

# Print non-numeric values found in each column
for col, non_numeric_vals in non_numeric_values.items():
    print(f"Non-numeric values in '{col}' column:", non_numeric_vals)
```
