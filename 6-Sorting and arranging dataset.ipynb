```python
import pandas as pd

# Specify the file path
excel_file = 'sales data.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file)
```

```python
# Remove duplicate rows based on 'Country'
df_no_duplicates = df.drop_duplicates(subset=['country'])
```

```python
# Sort DataFrame by the 'Country' column in ascending order
df_sorted = df_no_duplicates.sort_values(by='country', ascending=True)
```

```python
# Print the sorted data
df_sorted 
```

```python
# Set sorting by column name
sorted_device_df = df.sort_values(by='device_type', ascending=True)
```
