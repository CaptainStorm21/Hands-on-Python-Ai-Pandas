```python
import pandas as pd

# Replace 'your_file.xlsx' with the path to your Excel file
# If your Excel file is in the same directory as your Python script, you can simply specify the filename
extra_data = 'extra data.xlsx'
sales_data = 'sales data.xlsx'

# Load data from Excel file into a DataFrame
extra_data = pd.read_excel(extra_data)
sales_data = pd.read_excel(sales_data)
```

```python
extra_data.head()
```

```python
# Filter data where device_type is equal to 'PC'
filtered_df = sales_data[sales_data ['device_type'] == 'PC']
```

```python
# Filter data where device_type is equal to 'PC'
filtered_extra = extra_data[sales_data ['device_type'] == 'PC']
```

```python
# Give me a python code for data concat
```

```python
# Concatenating dataframes vertically
concatenated_df = pd.concat([filtered_df, filtered_extra], ignore_index=False)

# Displaying the concatenated dataframe
concatenated_df
```
