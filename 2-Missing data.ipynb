```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Specify the file path
excel_file = 'sales data.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file)

# Assuming you have already loaded your data into a DataFrame called df
# If not, you can load it using pd.read_excel(), pd.read_csv(), or another appropriate method

# Check for missing values
missing_values = df.isna().sum()

# Display missing values
print("Missing values in the data:")
print(missing_values)

# Display the head
df.head()
```

```python
# Assuming you have already loaded your data into a DataFrame called df
# If not, you can load it using pd.read_excel(), pd.read_csv(), or another appropriate method

# Specify the strategy for imputation (e.g., 'mean', 'median', 'most_frequent', etc.)
imputer = SimpleImputer(strategy='median')  # You can change the strategy as needed

# Fit the imputer to the data
imputer.fit(df)

# Impute missing values in the data
imputed_data = imputer.transform(df)

# Convert the imputed data back to a DataFrame (optional)
imputed_df = pd.DataFrame(imputed_data, columns=df.columns)

# Display the DataFrame with imputed values
print(imputed_df)
```

```python
# Assuming you have already loaded your data into a DataFrame called df
# If not, you can load it using pd.read_excel(), pd.read_csv(), or another appropriate method

# Define the strategy for imputation (e.g., 'mean', 'median', 'most_frequent', etc.)
imputer = SimpleImputer(strategy='most_frequent')  # You can change the strategy as needed

# Specify the columns with missing values that you want to impute
columns_with_missing_values = ['order_value_EUR', 'device_type']  # Replace with your column names

# Impute missing values in the specified columns
df[columns_with_missing_values] = imputer.fit_transform(df[columns_with_missing_values])

# Display the DataFrame with imputed values
print(df)
```
