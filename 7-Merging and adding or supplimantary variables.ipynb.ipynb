```python
import pandas as pd

# Replace 'your_file.xlsx' with the path to your Excel file
# If your Excel file is in the same directory as your Python script, you can simply specify the filename
excel_file = 'extra variable.xlsx'
sales_data = 'sales data.xlsx'

# Load data from Excel file into a DataFrame
extra_vars = pd.read_excel(excel_file)
sales_data = pd.read_excel(sales_data)
```

```python
# Merge the two dataframes on 'order_id'
merged_df = pd.merge(sales_data, extra_vars, on='order_id', how='inner')
```

```python
# Filter rows where 'Category' is 'Games' and 'Device_Type' is 'PC'
filtered_df = sales_data[(sales_data['category'] == 'Clothing') & (sales_data['device_type'] == 'PC')]

# Display the filtered data
print(filtered_df)
```
