```python
import pandas as pd

# Specify the file path
excel_file = 'sales data.xlsx'

# Load the Excel file into a DataFrame
df = pd.read_excel(excel_file)

# Display the DataFrame
print(df)
```

```python
# Display a number of rows you want to show 
df.head(20)
```

```python
# Identify duplicate rows based on all columns
duplicated_rows = df[df.duplicated()]

# Display duplicated rows
print ("Duplicated Rows: ")
print (duplicated_rows)
```

```python
# Remove duplicated Rows keeping the first accurence 
df_no_duplicates = df.drop_duplicates()

# Print the DataFrame without duplicates 
print ("Dataframe without Duplicates: ")
df_no_duplicates
```

```python
# Create a new variable to display a data without duplicates
clean_data = df_no_duplicates

# Print clean data
clean_data.head(20)
```
